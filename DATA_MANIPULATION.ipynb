**Joins**

```python
df1 = pd.DataFrame(  {"key" : ['A','B','C','D'],
                      "value" : [1,2,3,4] }  )
df2 = pd.DataFrame({"key" : ['B','D','D','E'],
                    "value" : np.random.randn(4)})
```

```python
pd.merge(df1,df2,on='key') # merge performs an INNER JOIN by default
# select * FROM df1 INNER JOIN df2 on df1.key = df2.key;
```

```python
pd.merge(df1,df2,on='key',how='left') # (  select * from df1 LEFT OUTER JOIN df2 on df1.key = df2.key )
```

```python
pd.merge(df1,df2,on='key',how='right') # ( SELECT * FROM df1 RIGHT OUTER JOIN df2 ON df1.key = df2.key;  )
```

```python
pd.merge(df1, df2, on="key", how="outer")
```

**update & delete ,Rank,if/else logic(where)**

```python
# update tips set tip = tip *2 where tip < 2 ;
tips.loc[tips['tip'] < 2 , "tip"] *= 2  #  (*= 2 is multiply and assign)
```

```python
tips.head(5)
```

```python
# DELETE FROM tips WHERE tip > 2;
''' In pandas we select the rows that should remain instead of deleting them: '''
tips = tips.loc[tips["tip"] <= 2]
```

```python
tips.head()
```

```python
''' SELECT * FROM tips
    ORDER BY tip DESC
    LIMIT 10 OFFSET 5 '''
tips.nlargest(10 + 5, columns="total_bill")
```

```python
tips.assign( rnk = tips.groupby(['day'])['total_bill'].rank(method ='first',ascending = False)
            ).query("rnk < 5").sort_values(['day','rnk']) 
```

```python
'''-- Oracle's ROW_NUMBER() analytic function
SELECT * FROM (
  SELECT
    t.*,
    ROW_NUMBER() OVER(PARTITION BY day ORDER BY total_bill DESC) AS rn
  FROM tips t
)
WHERE rn < 3
ORDER BY day, rn;'''
```
